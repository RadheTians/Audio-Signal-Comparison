```python
import pyaudio
import struct
```

```python
CHUNK = 1024
```

```python
CHANNELS = 1
```

```python
FORMAT = pyaudio.paInt16
```

```python
# samples per second
RATE = 48000
```

```python
# python connect to our mic
p = pyaudio.PyAudio()
```

```python
stream = p.open(
    format = FORMAT,
    channels = CHANNELS,
    rate = RATE,
    frames_per_buffer = CHUNK,
    input = True,
    output = True
)
```

```python
type(stream)
```

```python
data = stream.read(CHUNK)
data_int = struct.unpack(str(CHUNK*2) + "B" , data)
```

```python
len(data)
```
